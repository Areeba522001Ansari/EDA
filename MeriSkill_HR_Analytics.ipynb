# Rate and Number of the employees who works overtime

```python
# Calculate counts and percentages
counts = data['OverTime'].value_counts()
percentages = counts / len(data['OverTime']) * 100

# Create the count plot with enhanced features
plt.figure(figsize=(8, 8))
sns.set(style="whitegrid")
colors = sns.color_palette("YlGnBu", len(counts))
ax = sns.countplot(data=data, x='OverTime', palette=colors, edgecolor='0.2', linewidth=2)
plt.title('Employee Overtime Status', fontsize=18, fontweight='bold', color='#333333')
plt.xlabel('Overtime', fontsize=14, fontweight='bold', color='#333333')
plt.ylabel('Count', fontsize=14, fontweight='bold', color='#333333')

# Add counts, percentages, and numbers on the bars
for i, (count, percentage) in enumerate(zip(counts, percentages)):
    ax.text(i, count + 0.2, f'{count}\n{percentage:.1f}%', ha='center', fontsize=12, fontweight='bold', color='black')

# Remove y-axis ticks and spines for cleaner appearance
plt.yticks([])
sns.despine(left=True)

plt.show()
```

# Counting the number of the employee who are married and unmaried

```python
# Count the occurrences of each unique value in 'MaritalStatus'
counts = data['MaritalStatus'].value_counts()

# Calculate percentages
percentages = (counts / counts.sum()) * 100

# Create the count plot with enhanced features
plt.figure(figsize=(8, 6))
sns.set_style("whitegrid")
sns.barplot(x=percentages.index, y=percentages.values, palette='pastel')
plt.title('Marital Status Distribution', fontsize=18, fontweight='bold', color='#333333')
plt.xlabel('Marital Status', fontsize=14, fontweight='bold', color='#333333')
plt.ylabel('Percentage', fontsize=14, fontweight='bold', color='#333333')
plt.xticks(fontsize=12, rotation=45)  # Rotate x-axis labels for better visibility
plt.yticks(fontsize=12)
plt.ylim(0, 100)  # Set y-axis limit to represent percentages

# Add data labels on the bars
for i, p in enumerate(plt.gca().patches):
    plt.gca().annotate(f'{percentages.values[i]:.1f}%', (p.get_x() + p.get_width() / 2, p.get_height()),
                       ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=12)

plt.show()
```

# Checking the Educational background of the employees

```python

# Calculate counts and percentages
counts = data['EducationField'].value_counts()
percentages = counts / len(data['EducationField']) * 100

# Create the count plot with enhanced features
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
colors = sns.color_palette("pastel")
ax = sns.countplot(data=data, y='EducationField', palette=colors, edgecolor='0.2', linewidth=2)
plt.title('Education Field Distribution', fontsize=18, fontweight='bold', color='#333333')
plt.xlabel('Count', fontsize=14, fontweight='bold', color='#333333')
plt.ylabel('Education Field', fontsize=14, fontweight='bold', color='#333333')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add count and percentage labels on the bars
for i, (count, percentage) in enumerate(zip(counts, percentages)):
    ax.text(count + 0.2, i, f'{count} ({percentage:.1f}%)', va='center', fontsize=12, fontweight='bold')

# Add horizontal grid lines for a clean background
plt.gca().xaxis.grid(True, linestyle='--', linewidth=0.5)

plt.show()
```

# Age Distribution who does overtime

```python
# Create an interactive box plot with hover functionality
fig = px.box(data, x='OverTime', y='Age', title='Age Distribution Based on Overtime',
             labels={'Age': 'Age'},
             category_orders={'OverTime': ['Yes', 'No']})

# Customize hover text
fig.update_traces(hovertemplate='Age: %{y}<br>Overtime: %{x}')

# Customize the layout
fig.update_layout(title_x=0.5, title_y=0.9, title_font=dict(size=18),
                  xaxis_title='Overtime', yaxis_title='Age',
                  font=dict(family='Arial, sans-serif', size=12),
                  showlegend=False)

# Show the interactive plot
fig.show()
```
